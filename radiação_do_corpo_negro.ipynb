**Radiação do corpo negro**

Todos os corpos físicos emitem radiação térmica, Quando estão em equilíbrio térmico com o ambiente emitem tanta radiação  Quanto absorvem. Um corpo negro, absorve toda a radiação incidente nele Independentemente do comprimento de onda, Esta É a definição de um corpo negro.

```
import numpy as np
import matplotlib.pyplot as plt


# Constantes 

T1 = 3500 # K       Temperatura
T2 = 4000 # K       Temperatura
T3 = 4500 # K       Temperatura
pi = 3.14
c = 2.997925*10**(8)   # m/s  # velocidade da luz no vácuo
k = 1.381*10**(-23)    # J/k  # Constante de Boltzman


v = np.arange(0.1, 10000000000000, 1000000000)  # Frequência


# Definindo a função
def u1(v):
  u1 = (8*pi*1*k/c**3)*T1*v**2
  return u1

def u2(v):
  u2 = (8*pi*1*k/c**3)*T2*v**2
  return u2

def u3(v):
  u3 = (8*pi*1*k/c**3)*T3*v**2
  return u3

plt.xlabel('frequência')
plt.ylabel('Radiância espectral')
plt.plot(v, u1(v), 'k:', label = '4500 K', color = 'blue')
plt.plot(v, u2(v), 'k', label = '4000 K', color = 'yellow')
plt.plot(v, u3(v), 'k-.', label = '4500 K', color = 'orange')

plt.legend()
plt.show()
```

Essa equação se ajusta bem aos dados obtidos experimentalmente mas apenas em regiões de baixas frequências, essa equação não batia com os dados experimentais obtidos na época.

Em dezembro de 1900 Max Planck apresentou a sociedade de física alemã a teoria da lei de distribuição de energia do espectro normal, foi onde a mecânica quântica começou a ganhar forma, Plank criou um modelo onde os átomos das paredes do corpo negro estão em equilíbrio térmico com o campo de radiação, para auxiliar no tratamento estatístico dos dados ele assumiu que é energia viajava em pacotes Discreto e cada pacote tinha que ser um múltiplo inteiro de hv. A derivação da lei de radiação De Planck é:

```
import numpy as np
import matplotlib.pyplot as plt

# Constantes 
T1 = 4500 # K       Temperatura
T2 = 5000 # K       Temperatura
T3 = 5500 # K       Temperatura
T4 = 6000 # K       Temperatura
pi = 3.14
h = 6.62607*10**(-34)  # J s  # constante de Plank
c = 2.997925*10**(8)   # m/s  # velocidade da luz no vácuo
k = 1.381*10**(-23)    # J/k  # Constante de Boltzman

v = np.arange(0.1, 1000000000000000, 100000000000)  # comprimento de onda

def u1(v):
  u1 = (8*pi*h*v**3)/(c**2*(np.exp((h*v)/(k*T1))-1))
  return u1

def u2(v):
  u2 = (8*pi*h*v**3)/(c**2*(np.exp((h*v)/(k*T2))-1))
  return u2

def u3(v):
  u3 = (8*pi*h*v**3)/(c**2*(np.exp((h*v)/(k*T3))-1))
  return u3

def u4(v):
  u4 = (8*pi*h*v**3)/(c**2*(np.exp((h*v)/(k*T4))-1))
  return u4

plt.xlabel('frequência')
plt.ylabel('Radiância espectral')
plt.plot(v, u1(v), 'k:', label = '4500 K', color = 'blue')
plt.plot(v, u2(v), 'k', label = '5000 K', color = 'yellow')
plt.plot(v, u3(v), 'k-.', label = '5500 K', color = 'orange')
plt.plot(v, u4(v), 'k--', label = '6000 K', color = 'red')


plt.legend()
plt.show()
```
